```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
#defining equi-distant points in between -3 and 3
X = np.linspace(-3,3,71)
#calculating f(x) = x**2
Y = X**2
```

```python
#calculating X,Y( f(x) )
plt.plot(X,Y)

#defining points for demo
x = 2
dx = 0.3
dy = np.abs(x**2-(x+dx)**2)

#plotting the lines to create the triangle
plt.plot((x,x),(0,x**2))
plt.plot((x+dx,x+dx),(0,(x+dx)**2))
plt.plot((x,x+dx),(x**2,(x)**2))

#annotation
plt.text(x,x**2-dy/2,"dx",fontsize=12)
plt.text(x+dx,x**2+dy/2, " dy",fontsize=12)
```
